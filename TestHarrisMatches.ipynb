```python
from PIL import Image
from numpy import *
from pylab import *
import harris2
```

```python
im1 = array(Image.open('/content/drive/MyDrive/CIS3/Assignments/First/Images/TT22.jpg').convert('L'))
im2 = array(Image.open('/content/drive/MyDrive/CIS3/Assignments/First/Images/TT2.jpg').convert('L'))
```

```python
imshow(im1)
axis('off')
```

```python
imshow(im2)
axis('off')
```

```python
wid = 10
harrisim = harris2.compute_harris_response(im1,5)
filtered_coords1 = harris2.get_harris_points(harrisim,20,0.05)
d1 = harris2.get_descriptors(im1,filtered_coords1,wid)

harrisim = harris2.compute_harris_response(im2,5)
filtered_coords2 = harris2.get_harris_points(harrisim,20,0.05)
d2 = harris2.get_descriptors(im2,filtered_coords2,wid)
```

```python
harris2.plot_harris_corners(im1,filtered_coords1)
```

```python
harris2.plot_harris_corners(im2,filtered_coords2)
```

```python
print('Starting matching')
matches = harris2.match_twosided(d1,d2,0.8)
print(len(matches))
```

```python
print(matches)
len(matches)
```

```python
figure()
gray()
harris2.plot_matches(im1,im2,filtered_coords1,filtered_coords2,matches,False)
show()
```

```python
figure()
gray()
harris.plot_matches1(im1,im2,filtered_coords1,filtered_coords2,matches,False)
show()
```
